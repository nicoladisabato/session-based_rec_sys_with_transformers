# Making performance linechart for each model trained


# GRU4Rec

```python
with open('/content/drive/MyDrive/dataset_rees46/model_performances/gru_performance.txt', 'r') as f:
  lines = f.readlines()
```

```python
gru_ndcg_10 = []

for line in lines:
  # check if string present on a current line
  if line.find('ndcg_at_10') != -1:
      gru_ndcg_10.append(float(line.split('=')[-1]))

gru_ndcg_10
```

```python
gru_ndcg_20 = []

for line in lines:
  # check if string present on a current line
  if line.find('ndcg_at_20') != -1:
      gru_ndcg_20.append(float(line.split('=')[-1]))

gru_ndcg_20
```

```python
gru_recall_10 = []

for line in lines:
  # check if string present on a current line
  if line.find('recall_at_10') != -1:
      gru_recall_10.append(float(line.split('=')[-1]))

gru_recall_10
```

```python
gru_recall_20 = []

for line in lines:
  # check if string present on a current line
  if line.find('recall_at_20') != -1:
      gru_recall_20.append(float(line.split('=')[-1]))

gru_recall_20
```

# Creating the linechart

```python
days = []
for i in range(2,31):
  days.append(i)
```

```python
import matplotlib.pyplot as plt

plt.plot(days, gru_recall_10, label = "Recall@10", marker = 'o')
plt.plot(days, gru_recall_20, label = "Recall@20",  marker = 'o')
plt.title('GRU4Rec Recall')
plt.xlabel('days')
plt.ylabel('recall')
plt.legend()
plt.rcParams["figure.figsize"] = (10,6)
plt.show()

plt.plot(days, gru_ndcg_10, label = "NDCG@10", marker = 'o')
plt.plot(days, gru_ndcg_20, label = "NDCG@20",  marker = 'o')
plt.title('GRU4Rec NDCG')
plt.xlabel('days')
plt.ylabel('ndcg')
plt.legend()
plt.show()
```

```python
gru_eval_loss = []

for line in lines:
  # check if string present on a current line
  if line.find('eval_/loss') != -1:
      gru_eval_loss.append(float(line.split('=')[-1]))

gru_eval_loss
```

```python
plt.plot(days, gru_eval_loss, marker = 'o')
plt.title('Evaluation loss')
plt.xlabel('days')
plt.ylabel('loss')
plt.legend()
plt.ylim(5, 9)
plt.rcParams["figure.figsize"] = (10,6)
plt.show()
```
